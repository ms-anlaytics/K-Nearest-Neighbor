### Part II: Method Justification

B. Reasons for k-nearest neighbor (KNN) classification method:

B 1. KNN uses the most basic assumption when predicting, data clusters that are like one another will likely have the same outcomes. KNN must read through the entire dataset to find k nearest neighbors. Despite the first assumption of prediction, there are no other assumptions that need be met because KNN is a non-parametric algorithm meaning there are no other assumptions required. KNN does not build a model like regression instead it uses historical data to classify new data as a new data class to the nearest neighbor. KNN is designed to address multiclass problems and can be used for classification and regression tasks. KNN selects one hyperparameter to classify new data to and align the rest of the parameters. The KNN Euclidean Distance is the distance between points of classified parameters. (Christopher, 2021) 
<p/>
Since the KNN algorithm measures the probability that test data belongs to the classes of the ‘K’ training data and selects a class that holds the highest probability as an associated parameter (Christopher, 2021), data that is closely associated to the churn variable will be selected from the reset of the dataset. This will answer the the second point above and answers the question of how to find related variables more capable of classifying customer churn form the dataset. If the question is only choosing between customers who leave and customers who stay, represented by the churn variable, then the research question is answered by identifying k nearest neighbors to customer accounts that do churn. (Christopher, 2021)

<p/>

B 2. One assumption of KNN is the fact it is slow compared to other algorithms, because it learns from the historical data present to classify new data within the dataset. This means it will require more resources in memory than other algorithms because the classification step is iterative across the entire data set.

```python
# Import Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
# Ignore Warning Code
import warnings
warnings.filterwarnings('ignore')
```

```python
# import data set
dataset = pd.read_csv('churn_raw_data.csv')
dataset.head(1)
```

```python
# identify features
dataset.dtypes
```

Step 1. Remove unnecessary variables in the dataset:
<p/>
<code> 'Job','State','City','Timezone','Unnamed: 0','CaseOrder','Customer_id','Interaction','County','Zip','Lat', 'Lng','Population','Area','TimeZone','Email', 'Contacts'<code/> 


```python
# reducing the data into a collection of categorical and numeric data

indexes=['Job','State','City','Timezone','Unnamed: 0','CaseOrder','Customer_id','Interaction','County','Zip','Lat', 'Lng','Population','Area','TimeZone','Email', 'Contacts']
for x in indexes:
    if x in dataset.columns:
        dataset.drop(x, axis=1, inplace=True)
dataset.head(1)
```

Step 2. Drop duplicated data:

```python
#drop duplicates
dataset.drop_duplicates().reset_index(drop=True,inplace=True)
```

Step 3. Rename columns:

```python
# rename columns
dataset.rename(columns = {'item1':'Timely_Response','item2':'Timely_Fixes','item3':'Timely_Replacements','item4':'Reliability','item5':'Options','item6':'Respectful_Response','item7':'Courteous_Exchange','item8':'Active_Listening'}, inplace = True)
```

Step 4. Make numeric_data and category_data type lists

```python
# make numeric_data and category_data type lists
numeric_data = dataset.columns[dataset.dtypes != 'object']
category_data = dataset.columns[dataset.dtypes == 'object']
```

Step 5. Impute Null values with the mean:

```python
# replace null values with mean (Kumar, 2020)
dataset[numeric_data] = dataset[numeric_data].fillna(dataset.mean())
```

Step 6. Drop all rows with missing categorical data, as there is no way to impute it.

```python
dataset.dropna(inplace=True)
```

Verify no null values remain.

```python
dataset.info()
```

Step 7. Encode categorical variables.

```python
#7A.  Get one hot encoding of category_data columns 
one_hot = pd.get_dummies(dataset[category_data],drop_first=True)
```

```python
#7B. Drop columns of category_data from encoded data
dataset1 = dataset.drop(category_data,axis = 1)
```

```python
#7C. Join the encoded data
dataset1 = dataset1.join(one_hot)
```

```python
#Tidy up column names of encoded data. 
dataset1.columns=dataset1.columns.str.replace(' ', '_')
dataset1.columns=dataset1.columns.str.replace(',', '_')
dataset1.columns=dataset1.columns.str.replace('\'s', '')
dataset1.columns=dataset1.columns.str.replace('-', '')
dataset1.columns=dataset1.columns.str.replace('\(automatic\)', '_auto')
```

#### C4a. Univariate Analysis:
    
The univariate visualization graphs display a distribution plot and boxplot of the quartiles and distribution of outliers withing the variable

```python
## C4a. Univariate Analysis visual method:
import seaborn as sns

def univisual(x):
    sns.set(rc={"figure.figsize":(20, 5)})
    #x = 'Age'
    fig, ax = plt.subplots(nrows=1, ncols=2,  sharex=False, sharey=False)
    fig.suptitle(x, fontsize=20)
    ### distribution
    ax[0].title.set_text('distribution')
    variable = dataset1[x].fillna(dataset1[x].mean())
    breaks = np.quantile(variable, q=np.linspace(0, 1, 11))
    variable = variable[ (variable > breaks[0]) & (variable < 
                        breaks[10]) ]
    sns.distplot(variable, hist=True, kde=True, kde_kws={"shade": True}, ax=ax[0])
    des = dataset1[x].describe()
    ax[0].axvline(des["25%"], ls='--')
    ax[0].axvline(des["mean"], ls='--')
    ax[0].axvline(des["75%"], ls='--')
    ax[0].grid(True)
    des = round(des, 2).apply(lambda x: str(x))
    box = '\n'.join(("min: "+des["min"], "25%: "+des["25%"], "mean: "+des["mean"], "75%: "+des["75%"], "max: "+des["max"]))
    ax[0].text(0.95, 0.95, box, transform=ax[0].transAxes, fontsize=10, va='top', ha="right", bbox=dict(boxstyle='round', facecolor='white', alpha=1))

    ### boxplot 
    ax[1].title.set_text('outliers (log scale)')
    tmp_dtf = pd.DataFrame(dataset1[x])
    tmp_dtf[x] = np.log(tmp_dtf[x])
    tmp_dtf.boxplot(column=x, ax=ax[1])
    plt.show()
```

#### Data analysis:

D1. Import the cleaned data:

```python
#cleaned data
dataset2 = pd.read_csv('KNN_data.csv')
dataset2.drop('Unnamed: 0', axis=1, inplace=True)
```

D2. Divide data into X and y variables to correspond to predictor and responses variables.

```python
#divide data into X and y. Matrix = X and target = y
independent=list(dataset2.drop('Churn_Yes',axis=1).columns)
dependent='Churn_Yes'
X = dataset2[independent]
y = dataset2[dependent]
```

D3. Import machine learning packages to perform operations for splitting and testing in Scikit-learn.

```python
# Import machine learning packagesfrom Scikit-learn
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, classification_report
from sklearn.model_selection import cross_val_score, train_test_split
```

D3a. Create training and test sets

```python
#training and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.20, random_state = 1)
```

#### Preliminary   Results:

```python
#classification report
print(classification_report(y_test, y_pred))
```

D3e. Perform data standardization to normalize the data to prevent multicollinearity errors and reduce outlier errors, while increasing the precision of the dataset.

2. Split the treated data into training and test sets again.

```python
#training and test sets 
X_train_scaled, X_test_scaled, y_train_scaled, y_test_scaled = train_test_split(X, y, test_size = 0.2, random_state = 1)
```

3. Fit the train and test data to the "pipe" object.

4. Make a prediction from the new standardized data. 

```python
#predict with knn 
y_pred_scaled = pipe.predict(X_test_scaled)
```

#### Results:

7. Import Scikit-learn class to perform test of unseen data using a confusion matrix. The result will show the number of Type I and II errors the KNN algorithm makes with standardized data.  (How to Learn Machine Learning n.d.)

```python
#confusion matrix. (Buchanan 2021)
from sklearn.metrics import confusion_matrix

#visual method
def matrix(y_test, y_pred_scaled):
    cf_matrix = confusion_matrix(y_test, y_pred_scaled)
    print("\nTest confusion_matrix")
    sns.set(rc = {'figure.figsize':(8,8)})
    plt.xlabel('Predicted', fontsize=12)
    plt.ylabel('True', fontsize=12)

    group_names = ['True Neg', 'I False Positive', 'II False Negative', 'True Pos']
    group_counts = ["{0:0.0f}".format(value) for value in cf_matrix.flatten()]
    group_percentages = ["{0:.2%}".format(value) for value in cf_matrix.flatten()/np.sum(cf_matrix)]
    labels = [f"{v1}\n{v2}\n{v3}" for v1, v2, v3 in zip(group_names,group_counts,group_percentages)]
    labels = np.asarray(labels).reshape(2,2)
    sns.heatmap(cf_matrix, annot=labels, fmt='', cmap='Blues')
    
#display method 
matrix(y_test, y_pred_scaled)
```

### Part V: Data Summary and Implications

E.  Summarize of data analysis:




E1.  Accuracy and the area under the curve (AUC) analysis: Best KNN model parameters.

```python
#E1 (Christopher 2021)
# Import GridSearchCV for cross validation of model
from sklearn.model_selection import GridSearchCV

# Set up parameters grid
param_grid = {'n_neighbors': np.arange(1, 50)}

# Re-intantiate KNN for cross validation
knn = KNeighborsClassifier()

# Instantiate GridSearch cross validation
knn_cv = GridSearchCV(knn , param_grid, cv=5)

# Fit model to 
knn_cv.fit(X_train_scaled, y_train_scaled)
y_pred_scaled = knn_cv.predict(X_test_scaled)

# Print best parameters
print('E1.Result: Best KNN model parameters using cross validation: {}'.format(knn_cv.best_params_))

#accuracy_score vs best_score_
print('E1.Result: KNN model accuracy score: {:0.2f}'.format(accuracy_score(y_test_scaled, y_pred_scaled)))
print('E1.Result: KNN model best score: {}'.format(round(knn_cv.best_score_, 2)))
```

The accuracy of the KNN model can be summarized as the average of samples scores tested across the entire dataset, whereas the best score is the highest test score achieved among all samples tested with unseen data.

<p/>

The Area Under the Curve” (AUC) of “Receiver Characteristic Operator” (AUC-ROC) test reveals how well the KNN algorithm performed in classifying new data. Sensitivity demonstrates what portion of the positive class of the confusion matrix got correctly categorized.

<p/>

    Sensitivity = TP/TP+FN
<p/>

The False Negative Rate (FNR) demonstrates what portion of the positive class of the confusion matrix was misclassified.

<p/>

    FNR = FN/ TP+TN

<p/>


Specificity describes the true negative rate. 

<p/>


    Specificity = TN/TN+FP

<p/>


False Positive Rate (FPR) demonstrates what portion of the negative class in the confusion matrix was misclassed.

<p/>

    FPR = FP/TN+FP = 1 – Specificity

<p/>

Correctly classified data will likely have a higher TNR and a lower FPR score.  ROC plots the probability curve of TNR and FPR scores with variable threshold values to separate the signal from the noise.

<p/>

E2. AUC Score:

<ul>
<li>AUC = 1: the classifier can perfectly identify positive and negative class points.</li>
    
<li>AUC = 0: predicts the negative inverse of AUC =1.</li>
    
<li>0.5 &lt;AUC &lt;1: indicates a high probability that the KNN classifier can distinguish between the positive and negative class values. </li>
    
<li>AUC=0.5: the classifier cannot distinguish a difference between positive and negative classifiers.</li>
</ul>

<p/>

E4. ROC curve is interpreted with X and y axis.  A higher X value means there are more False positives than True negatives. A higher Y-axis indicates more True positives than False negatives. 


```python
#E4 ROC curve
from sklearn.metrics import roc_auc_score
from sklearn.metrics import roc_curve
from sklearn.metrics import roc_auc_score

# knn
model= KNeighborsClassifier(n_neighbors=4)

# fit model
model.fit(X_train, y_train)

# Compute predicted probabilities: y_pred_prob
pred_prob = knn_cv.predict_proba(X_test_scaled)

#cross validation
fpr1, tpr1, thresh = roc_curve(y_test, pred_prob[:,1], pos_label=1)

#predict probabilities
pred_prob = model.predict_proba(X_test_scaled)

#roc curve StandardScalr 
fpr2, tpr2, thresh = roc_curve(y_test, pred_prob[:,1], pos_label=1)

# roc curve Random Classifier 
random_probs = [0 for i in range(len(y_test))]
p_fpr, p_tpr, _ = roc_curve(y_test, random_probs, pos_label=1)
```

E4. Plot of the ROC curves of project data using Matplotlib. GridSearch cross validation is used in to make the green ROC curve and StandardScalar to form the red ROC curve. 

```python
#E4. plot the ROC curves (Bhandari 2016)
import matplotlib.pyplot as plt
plt.style.use('seaborn')

# plot method
def roc_plot():
    plt.plot(fpr1, tpr1, linestyle='--',color='green', label='KNN_cross validation')
    plt.plot(fpr2, tpr2, linestyle='--',color='red', label='KNN')
    plt.plot(p_fpr, p_tpr, linestyle='--', color='blue', label='Random Classifier')
    # title
    plt.title('ROC curve')
    # x label
    plt.xlabel('False Positive Rate')
    # y label
    plt.ylabel('True Positive rate')

    plt.legend(loc='best')
    plt.savefig('ROC',dpi=300)
    plt.show();
roc_plot()
```
